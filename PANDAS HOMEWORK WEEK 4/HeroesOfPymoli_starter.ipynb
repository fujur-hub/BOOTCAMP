```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
purchases = pd.read_csv("Resources/purchase_data.csv")
purchases.head()
```

## Player Count

* Display the total number of players


```python
#Count the number of unique player names
playerCnt = purchases.nunique()["SN"]
#Run basic calculations to obtain number of unique items, average price, etc.
playerDisplay = pd.DataFrame({"Total Players":[playerCnt]})
playerDisplay
```

## Purchasing Analysis (Total)

```python
#count unique items
uniqueItems = purchases["Item ID"].nunique()
#avg purchase price
avgPurchase = purchases["Price"].mean()
#total number of purchases
cntPurchase = purchases["Price"].count()
#total revenue
totPurchase = purchases["Price"].sum()

#new dataframe
purchaseTable = pd.DataFrame({"Number of Unique Items": uniqueItems,
                              "Average Purchase Price": avgPurchase,
                              "Number of Purchases": cntPurchase,
                              "Total Revenue": [totPurchase]})

#adds the formatted average purchase price in place of the unformatted average purchase price column
purchaseTable ["Average Purchase Price"] = purchaseTable["Average Purchase Price"].map("${:,.2f}".format)

#adds the formatted total revenue in place of the unformatted total revenue column
purchaseTable ["Total Revenue"] = purchaseTable["Total Revenue"].map("${:,.2f}".format)

purchaseTable
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#group by gender
genderData = purchases.groupby("Gender") 

#count screen names by gender
genderCnt = genderData.nunique()["SN"]

#calculates the percentage of players by gender
playerPercentage = genderCnt / playerCnt *100 

#creates the table for the data.
genderDemographics = pd.DataFrame({"Percentage of Players" : playerPercentage,
                                   "Total Count": genderCnt})

#formats table so that there is no index name in  corner
genderDemographics.index.name = "Gender"

#formats table so that total count is in decending order and so that there are 2 decimal places for the percentage
genderDemographics.sort_values(["Total Count"], ascending = False).style.format({"Percentage of Players":"{:.2f}"})
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages--ugh this took too long
ageBins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
groupHeaders = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Segment and sort age values into bins established above
purchases["Age Group"] = pd.cut(purchases["Age"],ageBins, labels=groupHeaders)
purchases

# Create new data frame with the added "Age Group" and group it
ageGroup = purchases.groupby("Age Group")

#counts the total number of purchases by age group
totAgeCnt = ageGroup["Purchase ID"].count()

# Obtain average purchase price by age group 
purchaseAgeAvg = ageGroup["Price"].mean()

# Calculate total purchase value by age group 
purchaseAgeTot = ageGroup["Price"].sum()

# Calculate the average purchase per person in the age group 
avgPurchaseAge = purchaseAgeTot/totAgeCnt

# Create data frame with obtained values
age_demographics = pd.DataFrame({"Purchase Count": totAgeCnt,
                                 "Average Purchase Price": purchaseAgeAvg,
                                 "Total Purchase Value":purchaseAgeTot,
                                 "Average Purchase Total per Person": avgPurchaseAge})

# Format the data frame with no index name in the corner
age_demographics.index.name = "Age Group"

# Format with currency style
age_demographics.style.format({"Average Purchase Price":"${:,.2f}",
                               "Total Purchase Value":"${:,.2f}",
                               "Average Purchase Total per Person":"${:,.2f}"})
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#group data by screen names
spenderData = purchases.groupby("SN")

#counts the total number of purchases by name
topSpenderCnt = spenderData["Purchase ID"].count()

#calculates the average purchase price by name
topAvgPurchase = spenderData["Price"].mean()

#calculate purchase total
topPurchaseTot = spenderData["Price"].sum()

#creates the table for the data
topSpenders = pd.DataFrame({"Purchase Count": topSpenderCnt,
                            "Average Purchase Price": topAvgPurchase,
                            "Total Purchase Value": topPurchaseTot})

#formats table so that the total purchase value is in decending order only storing the top 5 spenders
formatSpenders = topSpenders.sort_values(["Total Purchase Value"], ascending = False).head()

# Format with currency style
formatSpenders.style.format({"Purchase Count": "${:,.2f}",
                             "Average Purchase Price": "${:,.2f}",
                             "Total Purchase Value": "${:,.2f}"})
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#create new data frame with items related information
items = purchases[["Item ID", "Item Name", "Price"]]

#group item data by item id and item name
itemInfo = items.groupby(["Item ID", "Item Name"])

#count number of times an item has been purchased
itemCnt = itemInfo["Item Name"].count()

#calculate the purchase value per item
purchaseVal = itemInfo["Price"].sum()

#obtains the individual item price
itemPrice = purchaseVal/itemCnt

#creates the table for the data
mostPopItems = pd.DataFrame({"Purchase Count": itemCnt,
                             "Item Price": itemPrice,
                             "Total Purchase Value":purchaseVal})

#formats table so that the purchase purchase is in decending order only storing the top 5 most popular items 
popFormat = mostPopItems.sort_values(["Purchase Count"],ascending = False).head()

# Format with currency style
popFormat.style.format({"Item Price":"${:,.2f}",
                        "Total Purchase Value":"${:,.2f}"})
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order

```python
#formats table so that the purchase purchase is in decending order and only storing the top 5 most profitable items
popFormat = mostPopItems.sort_values(["Total Purchase Value"],ascending = False).head()

# Format with currency style
popFormat.style.format({"Item Price":"${:,.2f}",
                        "Total Purchase Value":"${:,.2f}"})
```
